# 10 — Geometry Overlay Comparison Prototype

This notebook placeholder documents how reference geometry and user-photo landmarks will be compared after the photo quality gate accepts an image.

```python
reference = {'body_length_height': 1.105, 'chest_depth_height': 0.50, 'head_length_height': 0.36, 'muzzle_skull': 0.50}
user_estimate = {'body_length_height': 1.14, 'chest_depth_height': 0.48, 'head_length_height': 0.34, 'muzzle_skull': 0.61}
for key, ref in reference.items():
    delta = (user_estimate[key] - ref) / ref
    print(key, round(delta * 100, 2), '%')
```
