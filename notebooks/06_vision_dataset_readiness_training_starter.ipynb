# Vision Dataset Readiness & Training Starter

This notebook explains the first practical step before training the visual Cane Corso model. It checks whether the image dataset is ready for neural training.


## Goal

The visual ML pipeline should learn three things in order:

1. Is the photo suitable for comparison?
2. What type of photo is it?
3. Is the visible dog visually close to the Cane Corso reference type?

Training must not start until the dataset has licensed images, balanced classes and reliable labels.


```python
from pathlib import Path
import pandas as pd

labels_path = Path('../data/images/labels/sample-image-labels.csv')
labels = pd.read_csv(labels_path)
labels.head()
```

```python
labels.groupby(['breed_label', 'comparison_readiness']).size().reset_index(name='count')
```

## Future training approach

When enough images are collected, the first training experiment can use transfer learning:

- MobileNet / EfficientNet / ResNet backbone
- image augmentation
- train/validation/test split
- metrics: accuracy, precision, recall, F1, confusion matrix

The first model should be a Photo Readiness Classifier. The Breed Visual Classifier should come after the photo gate is reliable.


## Safe output language

The final app should say **Visual Cane Corso Match**, not breed purity. A photo cannot prove pedigree, registration or genetic origin.
