## Safe formula

`match = gate(readiness) × (0.35 × classifier + 0.30 × embedding_similarity + 0.25 × geometry + 0.10 × confidence_adjustment)`


```python
def gated_visual_match(readiness, classifier, embedding_similarity, geometry, confidence_adjustment):
    if readiness == 'rejected':
        return None
    score = 0.35 * classifier + 0.30 * embedding_similarity + 0.25 * geometry + 0.10 * confidence_adjustment
    if readiness == 'limited':
        score *= 0.85
    return round(max(0.0, min(1.0, score)) * 100, 2)

gated_visual_match('accepted', 0.78, 0.74, 0.81, 0.88)
```

## Boundary

The output is visual similarity only. It does not prove pedigree, breed purity, genetic origin or official status.
