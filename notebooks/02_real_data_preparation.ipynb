## Check Whether the Raw Dataset Exists Locally

This check should not fail the notebook.

If the file is missing, it means the raw dataset has not been downloaded yet.


```python
if expected_raw_zip.exists():
    print('Raw dataset ZIP found locally.')
else:
    print('Raw dataset ZIP is not available locally yet.')
    print('Download instructions are documented in docs/real_data_download_instructions.md')
```

## Inspect ZIP Contents

When the raw ZIP file is downloaded locally, this section will list its contents.

For now, the code is safe and will only run if the file exists.


```python
if expected_raw_zip.exists():
    with zipfile.ZipFile(expected_raw_zip, 'r') as zip_file:
        files = zip_file.namelist()
        print('Files inside ZIP:')
        for file_name in files[:20]:
            print('-', file_name)
else:
    print('Skipping ZIP inspection because the raw dataset is not downloaded yet.')
```
