# 05 — Breed Classifier Training Plan

This notebook defines the planned neural vision workflow for the Cane Corso Growth Geometry Lab.

The goal is not to prove breed purity from a photo. The goal is to train visual models that can support an explainable **Visual Cane Corso Match** signal.

## Pipeline

1. Photo readiness model — accepted / limited / rejected.
2. Photo type classifier — side body / front body / head profile / head front / unsuitable.
3. Breed visual classifier — Cane Corso vs similar and negative classes.
4. Visual similarity model — image embedding compared with curated Cane Corso reference examples.
5. Geometry + vision fusion — combine classifier, similarity and reference-geometry signals.
