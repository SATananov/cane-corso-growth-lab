# Visual Match Result Contract

This notebook documents the final result contract for the future visual Cane Corso similarity model.

## Formula

`match = gate(readiness) × (0.35 × classifier + 0.30 × embedding + 0.25 × geometry + 0.10 × confidence)`

```python
weights = {
    "classifier": 0.35,
    "embedding": 0.30,
    "geometry": 0.25,
    "confidence": 0.10,
}
sum(weights.values())
```

```python
def visual_match(readiness, classifier, embedding, geometry, confidence):
    if readiness == "rejected":
        return None
    score = (
        0.35 * classifier
        + 0.30 * embedding
        + 0.25 * geometry
        + 0.10 * confidence
    )
    if readiness == "limited":
        score *= 0.88
    return round(score * 100, 1)

visual_match("accepted", 0.86, 0.82, 0.78, 0.90)
```

## Safety boundary

The score is a visual similarity signal only. It does not prove breed purity, pedigree, genetic origin, health or official registration.
